```python
%load_ext tensorboard
import os
from glob import glob
import pandas as pd
import matplotlib.pyplot as plt

import numpy as np
import tensorflow as tf
import datetime
from tensorflow import keras
from tensorflow.keras import layers
```

```python
# Get list of all the sentences
path = r"D:\\MinHyuk\\Hand Sign Recognition\\2021_SLT_prj\\DATA\\20210302CTC실험"
all_files = glob(path+"\\*.csv")

sentences = []
stc_seq = [[11, 13, 10, 7, 12],[11, 15, 10, 8, 1],[11, 9, 4, 8, 1],[3, 15, 4, 7, 2],[3, 15, 0, 8, 2],[11, 14, 6, 5, 2]]
labels = []
for filename in all_files:
    df = np.array(pd.read_csv(filename, header=None), dtype='float32')
    sentences.append(df)
    
    temp = filename.split(os.path.sep)[-1].split(".csv")[0]
    temp = temp.split('-')[0]
    temp = temp[-1]
    labels.append(np.array(stc_seq[int(temp)-1]))

for i, stc in enumerate(sentences):
    if i==0:
        stacked_sentences = stc
    else:
        stacked_sentences = np.vstack((stacked_sentences, stc))

sentence_mean = np.mean(stacked_sentences, axis=0)
sentence_std = np.std(stacked_sentences, axis=0)
for i, stc in enumerate(sentences):
    tmp = (stc - sentence_mean)/sentence_std
    #tmp = tf.ragged.constant(tmp)
    sentences[i] = tmp

#max_stc_len = max([len(stc) for stc in sentences])
#sentences = tf.ragged.constant(sentences)
#sentences = keras.preprocessing.sequence.pad_sequences(sentences, maxlen=max_stc_len, dtype='float32', padding='post', value=0)
```

```python
def split_data(sentences, labels, train_size=0.9, shuffle=True):
    # 1. Get the total size of the dataset
    size = len(sentences)
    # 2. Make an indices array and shuffle it, if required
    indices = np.arange(size)
    if shuffle:
        np.random.shuffle(indices)
    # 3. Get the size of training samples
    train_samples = int(size * train_size)
    # 4. Split data into training and validation sets
    x_train, y_train = sentences[indices[:train_samples]], labels[indices[:train_samples]]
    x_valid, y_valid = sentences[indices[train_samples:]], labels[indices[train_samples:]]
    
    return x_train, x_valid, y_train, y_valid

x_train, x_valid, y_train, y_valid = split_data(np.array(sentences), np.array(labels))
x_train = tf.ragged.constant(x_train)
x_valid = tf.ragged.constant(x_valid)

def encode_dict(stc, label):
    #stc = tf.ragged.constant(stc, dtype=tf.float32)
    #stc = tf.RaggedTensor.from_tensor(stc, padding=0)
    #stc = tf.convert_to_tensor(stc, dtype=tf.float32)
    label = tf.convert_to_tensor(label, dtype=tf.int32)
    return {"sensor": stc, "label": label}
```

```python
train_dataset = tf.data.Dataset.from_tensor_slices((x_train, y_train))
train_dataset = train_dataset.map(encode_dict, num_parallel_calls=tf.data.experimental.AUTOTUNE
                                 ).shuffle(100).batch(1).prefetch(buffer_size=tf.data.experimental.AUTOTUNE)

validation_dataset = tf.data.Dataset.from_tensor_slices((x_valid, y_valid))
validation_dataset = validation_dataset.map(encode_dict, num_parallel_calls=tf.data.experimental.AUTOTUNE
                                           ).shuffle(100).batch(1).prefetch(buffer_size=tf.data.experimental.AUTOTUNE)
```

```python
class CTCLayer(layers.Layer):
    def __init__(self, name=None):
        super().__init__(name=name)
        self.loss_fn = keras.backend.ctc_batch_cost
        
    def call(self, y_true, y_pred):
#         y_pred = y_pred.to_tensor()
        # compute the training-time loss value and add it
        # to the layer using 'self.add_loss()'
        batch_len = tf.cast(tf.shape(y_true)[0], dtype="int64") # shape의 [0] 차원은 batch_size를 의미함.
        input_length = tf.cast(tf.shape(y_pred)[1], dtype="int64") # 매 timestep마다 output 출력하므로 y_pred의 길이가 input의 길이와 동일.
        label_length = tf.cast(tf.shape(y_true)[1], dtype="int64") # 실제 출력되어야 할 문장의 길이. 이 예시 경우엔 5.
        
        input_length = input_length * tf.ones(shape=(batch_len, 1), dtype="int64") # input_length를 batch_size만큼 복제함.
        label_length = label_length * tf.ones(shape=(batch_len, 1), dtype="int64") # label_length를 batch_size만큼 복제함.
        
        loss = self.loss_fn(y_true, y_pred, input_length, label_length)
        self.add_loss(loss)
        
        # At test time, just return the computed predictions
        return y_pred
    
def build_model():
    with tf.device('/gpu:0'):
        # inputs to the model
        input_sensor = layers.Input(
            shape=(None,10), name="sensor", dtype="float32", ragged=True
        )
        #x = layers.Embedding(input_dim=10, output_dim=64, mask_zero=True)(input_sensor)
        #x = layers.Masking(mask_value=0.0, input_shape=(max_stc_len,10))(input_sensor)
        
        labels = layers.Input(name="label", shape=(None,), dtype="float32")
        
        # First conv block
        x = layers.Conv1D(
            filters=64,
            kernel_size=20,
            #strides=20,
            activation='relu',
            kernel_initializer="he_normal",
            padding="same",
            name="Conv1",
        )(input_sensor.to_tensor())
        x = layers.MaxPooling1D(2, name="pool1")(x)
        
        # Second conv block
        x = layers.Conv1D(
            filters=64,
            kernel_size=3,
            activation="relu",
            padding="same",
            name="Conv2",
        )(x)
        #x = layers.MaxPooling1D(2, name="pool2")(x)
        
        #x = layers.Reshape((-1,64))(x)
        x = layers.Dense(64, activation="relu", name="dense1")(x)
        x = layers.Dropout(0.2)(x)
        
        # Unidirectional LSTM
        x = layers.LSTM(128, return_sequences=True, dropout=0.25)(x)
        x = layers.LSTM(64, return_sequences=True, dropout=0.25)(x)
        
        # Bidirectional LSTM
        #x = layers.Bidirectional(layers.LSTM(128, return_sequences=True, dropout=0.25))(x)
        #x = layers.Bidirectional(layers.LSTM(64, return_sequences=True, dropout=0.25))(x)

        x = layers.TimeDistributed(layers.Dense(16 + 1, activation="softmax"), name="dense2")(x)
        
        # Add CTC layer for calculation CTC loss at each step
        output = CTCLayer(name="ctc-loss")(labels, x)
        
        # Define the model
        model = keras.models.Model(
            inputs=[input_sensor, labels], outputs=output, name="test_SLT"
        )
        
        # Optimizer
        opt = keras.optimizers.Adam()
        
        # Compile the model and return
        model.compile(optimizer=opt)
        return model
    
# Get the model
model = build_model()
model.summary()
```

```python
epochs = 200
early_stopping_patience = 20
# Add early stopping
early_stopping = keras.callbacks.EarlyStopping(
    monitor="val_loss", patience=early_stopping_patience, restore_best_weights=True
)

log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)

# Train the model
history = model.fit(
    train_dataset,
    validation_data=validation_dataset,
    epochs=epochs,
    callbacks=[early_stopping, tensorboard_callback]
)
```
